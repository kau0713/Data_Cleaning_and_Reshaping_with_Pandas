# Data_Cleaning_and_Reshaping


```python
# Import Library 
import pandas as pd
import numpy as np
```

```python
# Import So Much Candy Data
candy = pd.read_excel("candyhierarchy2017.xlsx") 
```

```python
candy.head() 
```

```python
# Check the names of the Columns
candy.columns
```

```python
candy.shape
```

```python
# Remove in missing data
candy.drop_duplicates(inplace=True)
candy.shape
```

```python
# Filter out missing data
candy.dropna(axis=0,how='all',inplace=True) # Drop the Nan value 

# Also dropping the columns with too many records are reported as Nan (>50%)
candy.dropna(thresh=candy.shape[0]*0.5,axis=1,inplace=True)
# Dropping the rows with too many records are reported as Nan (>50%)
candy.dropna(thresh=candy.shape[1]*0.5,axis=0, inplace=True)

candy.shape
```

```python
candy.columns
```

```python
# Rename some of the columns

new_columns_dict={
    'Q6 | Anonymous brown globs that come in black and orange wrappers\t(a.k.a. Mary Janes)':'Q6 | Mary Janes',
    'Q6 | Any full-sized candy bar':'Q6 | Full-sized candy bar',
    'Q6 | Vials of pure high fructose corn syrup, for main-lining into your vein':'Q6 | Pure high fructose corn syrup'
}
candy.rename(columns=new_columns_dict, inplace=True)
```

```python
# Remove the 'Q6' for all the candies name 
candy_names = {x:x.split('Q6 | ')[1] for x in candy.columns if x.startswith('Q6 | ')}
candy.rename(columns=candy_names,inplace=True)
candy.reset_index(inplace=True)
```

```python
candy.columns
```

```python
# Check for the missing values 
candy.isnull().sum()
```

```python
# Chpater 8
# Create hierarchical index

# Create the hierachial data with 'Gender' and 'Age'
candy.set_index(['Q2: GENDER', 'Q3: AGE'], inplace=True)
candy.head(10)
```

```python
# Reshaping Data 

# Create new index for Candy 
new_index = pd.Index(list(candy_names.values()),name='Candy')
candy_new = candy.reindex(columns=new_index)
```

```python
# Using the stack() to reshape 
candy_3 = candy_new.stack().reset_index().rename(columns={0: 'Enjoyment'})
candy_3[:10]
```

```python
candy.reset_index(inplace=True)
```

```python
# Chapter 10 
# Grouping with Dicts/Series
```

```python
# Trying to group the data by Country and City
candy.dropna(subset=['Q4: COUNTRY','Q5: STATE, PROVINCE, COUNTY, ETC'],inplace=True)

mapping = {
    'Q4: COUNTRY': 'COUNTRY',
    'Q5: STATE, PROVINCE, COUNTY, ETC': 'LOCATION'
}

candy_g = candy.groupby(mapping, axis=1)
candy_g
     
```
